```python
import pandas as pd
# 파일을 올바른 인코딩으로 읽어오기 (예: UTF-8)
df = pd.read_csv('floor.csv', encoding='euc-kr')
df
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt

grouped_data = df.groupby(['자치구']).sum()
colors = ['#f94144', '#f3722c', '#f8961e','#f9844a','#f9c74f','#90be6d','#43aa8b','#4d908e','#4d908e','#577590','#277da1']
explode = [0.14 if typ == '횡단중' else 0.04 for typ in grouped_data.index]

grouped_data['설치수량(조)'].plot(title = "서울시 바닥형 보행신호등 보조장치 설치현황", kind='bar',figsize = (15,5), rot=45,color=colors, alpha=0.5)
plt.show()
```
