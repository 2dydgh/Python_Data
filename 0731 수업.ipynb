```python
#https://https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105
# 분야별 주요뉴스 15개 수집
# 수집한 내용은 article.csv 파일에 저장
# 제목,공백의 수, 링크
# csv 라이브러리 사용하지 않고, open 함수와 write 함수를 이용해서 진행
```

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent


ua = UserAgent()
headers = {
    "User-Agent": ua.random
}

res = requests.get("https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105", headers=headers)
bs = BeautifulSoup(res.text,'html.parser')
area = bs.select_one(".classfy")
elem = area.select("a")  

f = open("article.csv","w",encoding="utf-8")
for e in elem:
    title = e.text.replace("," , "-")
    count = title.count(" ")
    link = e.attrs.get("href","")
    f.write(f"{title},{count},{link}\n")
f.close()

```

```python
# https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC
# 수집한 내용은 search.csv 파일에 저장
# 제목,요약,날짜
# csv 라이브러리 사용하지 않고, open 함수와 write 함수를 이용해서 진행

import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

ua = UserAgent()
headers = {
    "User-Agent": ua.random
}

res = requests.get("https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC", headers=headers)
bs = BeautifulSoup(res.text,'html.parser')
    
area = bs.select_one(".lst_total")
lis = area.select("._svp_item")
f = open("search.csv","w",encoding="utf-8")
for li in lis:
    title = li.select_one(".total_tit")
    title = title.text
    print(title)
    
    summary = li.select_one(".dsc_txt")
    summary = summary.text
    print(summary)
    
    date = li.select_one(".date")
    if not date:
        date = li.select_one(".sub_time")
    date = date.text
    print(date)
    f.write(f"{title},{summary},{date}\n")

```

```python
#tds = area.select("td들 가져오는 코드")
#tds : 리스트
#tds[0] : 가격
#tds[1] : 변동가격
#tds[2] : 변동률

import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

ua = UserAgent()
headers = {
    "User-Agent": ua.random
}

res = requests.get("https://finance.naver.com/", headers=headers)
bs = BeautifulSoup(res.text,'html.parser')

tbody = bs.select_one("#_topItems1")
trs = tbody.select("tr")
for tr in trs:
    name = tr.select_one("th > a")  # th 바로 밑에 a
    name = name.text
    
    tds = tr.select("td")
    price = tds[0].text.strip()
    diff = tds[1].text.strip()
    rate = tds[2].text.strip()

    print(name,price,diff,rate)

```
