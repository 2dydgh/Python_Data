```python
# <숫자 맞히기 스무고개 게임>
    # 프로그램 실행시 랜덤으로 1~100사이의 숫자를 생성
    # 10번의 기회가 있고, 10번동안 숫자를 입력 받아 숫자를 맞추는 게임
    # 정답이 아닐 때마다 up/ down 을 출력
    # 정답을 맞추면 "정답"을 출력
    # 10번의 기회동안 정답을 못 맞췄다면 "실패"를 출력
    
import random 

computer = random.randint(1, 100)

for i in range(10):
    guess =input("숫자를 입력하세요: ")
    guess = int(guess)
    
    if guess > computer:
        print("down")
    elif guess < computer:
        print("up")
    else:
        print("정답")
        break
         
# 여기서 i ==9 는 없어도 된다! 애초에 정답이 아니면 저길 빠져나오지 못함!
if i ==9 and guess != computer: 
    print("실패!")
```

```python
# <숫자 맞히기 스무고개 게임>
    # 프로그램 실행시 랜덤으로 1~100사이의 숫자를 생성
    # 10번의 기회가 있고, 10번동안 숫자를 입력 받아 숫자를 맞추는 게임
    # 정답이 아닐 때마다 up/ down 을 출력
    # 정답을 맞추면 "정답"을 출력
    # 10번의 기회동안 정답을 못 맞췄다면 "실패"를 출력
    
    # +)추가 
    # 입력값이 숫자가 아니거나 1~100사이의 숫자가 아닐 경우
    # 기회를 차감하지 않도록 변경
    # -> while문 써야함
import random 

computer = random.randint(1, 100)

chance =10
while chance > 0:
    guess = input("숫자를 입력하세요: ")
    
    if not guess.isdigit():
        print("다시 입력하세요.")
        print(chance)
        continue
        
    guess = int(guess)
    if not(1<= guess <=100):
        print("다시 입력하세요.")
        print(chance)
        continue
        
    if guess > computer:
        print("down")
        print(chance)
    elif guess < computer:
        print("up")
        print(chance)
    else:
        print("정답")
        break
    chance = chance -1
    
        
if  guess != computer: 
    print("실패!")
```

```python
# 함수로 만들기, 메뉴 만들기

# <숫자 맞히기 스무고개 게임>
    # 프로그램 실행시 랜덤으로 1~100사이의 숫자를 생성
    # 10번의 기회가 있고, 10번동안 숫자를 입력 받아 숫자를 맞추는 게임
    # 정답이 아닐 때마다 up/ down 을 출력
    # 정답을 맞추면 "정답"을 출력
    # 10번의 기회동안 정답을 못 맞췄다면 "실패"를 출력
    
    # +)추가 
    # 입력값이 숫자가 아니거나 1~100사이의 숫자가 아닐 경우
    # 기회를 차감하지 않도록 변경
    # -> while문 써야함
    
    # +)함수로 만들어서 
    # 1.누르면 게임 시작 2.누르면 게임 종료
import random 

def start_game(username):

    computer = random.randint(1, 100)

    chance =10
    while chance > 0:
        guess = input("숫자를 입력하세요: ")

        if not guess.isdigit():
            print("다시 입력하세요.")
            print(chance)
            continue

        guess = int(guess)
        if not(1<= guess <=100):
            print("다시 입력하세요.")
            print(chance)
            continue

        if guess > computer:
            print("down")
            print(chance)
        elif guess < computer:
            print("up")
            print(chance)
        else:
            print(f"{username} 정답!")
            break
        chance = chance -1


    if  guess != computer: 
        print(f"{username} 실패!")

        
username = input("이름을 입력하세요: ")        
while True:
        
    print("1. 게임 시작")
    print("2. 게임 종료")
    menu = input("메뉴 선택: ")
    if menu == "1":
        start_game(username)
    elif menu == "2":
        break

```
