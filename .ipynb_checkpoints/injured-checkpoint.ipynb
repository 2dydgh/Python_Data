```python
import pandas as pd
# 파일을 올바른 인코딩으로 읽어오기 (예: UTF-8)
df = pd.read_csv('injured.csv', encoding='euc-kr')
df
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt

grouped_data = df.groupby(['사고유형별']).sum()
# 파이 차트의 색상 설정
colors = ['#023047', '#219ebc', '#8ecae6','#fb8500','#ffb703']
explode = [0.14 if typ == '횡단중' else 0.04 for typ in grouped_data.index]

grouped_data['부상자수 (명)'].plot(title = "2022년 부상자수 통계", kind='pie',autopct = '%1.0f%%',
                        shadow = 'True' ,colors=colors, explode = explode, startangle=135)
plt.show()
```
