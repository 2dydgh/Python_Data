```python
import pandas as pd

df = pd.read_excel(io='disabled.xlsx', sheet_name = 'data')
df
```

```python
# 삭제할 열 이름 리스트
columns_to_drop = ['2018.1', '2018.2', '2019.1', '2019.2', '2020.1', '2020.2', '2021.1', '2021.2', '2022.1', '2022.2']

# 열 삭제
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.drop(['장애유형별(2)'], axis=1, inplace=True)
df
```

```python
df.drop(list(range(5,7)), axis=0)
```

```python
df = df.rename(columns = {
    '장애유형별(1)' :'장애유형', '2018' : '2018년','2019' : '2019년', '2020' : '2020년','2021' : '2021년', '2022' : '2022년'})
df
```

```python
df.drop(0, axis=0, inplace=True)
df
```

```python
df.drop(1, axis=0, inplace=True)
df
```

```python
df
```

```python
df.drop(list(range(3,5)), axis=0, inplace = True)
```

```python
df.drop(16, axis=0 ,inplace = True)
df
```

```python
# 특정 범위의 행 데이터를 선택하여 합계 계산
sum_values = df.iloc[6:16, 1:].sum()

# 새로운 행 생성
new_row = pd.DataFrame({'장애유형': ['기타'],  # 행 이름
                        '2018년': [sum_values['2018년']],
                        '2019년': [sum_values['2019년']],
                        '2020년': [sum_values['2020년']],
                        '2021년': [sum_values['2021년']],
                        '2022년': [sum_values['2022년']]})

# 새로운 행 추가하여 데이터프레임 합치기
df = pd.concat([df, new_row], ignore_index=True)
df
```

```python
df.drop(list(range(4,14)), axis=0,inplace=True)
```

```python
df
```

```python
df.reset_index()
```

```python
df.reset_index(drop=True, inplace = True)
df
```

```python
df.set_index('장애유형', inplace = True)
df
```

```python
df['평균'] = df.mean(axis=1)
```

```python
df
```

```python
df.drop(['2018년','2019년','2020년', '2021년', '2022년'], axis=1, inplace=True)
df
```

```python
df.plot(kind = 'bar')
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt

grouped_data = df.groupby(['장애유형']).sum()
# 파이 차트의 색상 설정
colors = ['#4D96FF', '#FFD93D', '#6BCB77', '#6F69AC', '#FF6B6B']
grouped_data['평균'].plot(title = "5년간 장애 유형별 분류", kind='pie',autopct = '%1.0f%%',
                        shadow = 'True', explode = (0.05,0.3,0.05,0.05,0.05) ,colors=colors)
plt.show()
```
