```python
# selenium을 이용해서 멜론 차트 가져오기
# diff도 추가하기 -> 태그.get_attribute("")를 통해 속성값을 가져올 수 있음

import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pymysql

db = pymysql.connect(host="localhost", port=3306, user="root", password = "1234", db = "sba")
cursor = db.cursor() 

service = Service(ChromeDriverManager().install())
d = webdriver.Chrome(service=service)

try:
    d.get("https://www.melon.com/chart/index.htm?dayTime=2023080115")
    
    area = d.find_element(By.CSS_SELECTOR, ".service_list_song")
    trs = area.find_elements(By.CSS_SELECTOR, "table>tbody>tr")

    for tr in trs:
        rank = tr.find_element(By.CSS_SELECTOR, ".rank")
        title = tr.find_element(By.CSS_SELECTOR,".ellipsis.rank01 a")
        singer = tr.find_element(By.CSS_SELECTOR,".ellipsis.rank02 a")
        album = tr.find_element(By.CSS_SELECTOR,".ellipsis.rank03 a")
        like = tr.find_element(By.CSS_SELECTOR,".like .cnt")
        # point! : 좋아요는 숫자가 아니다. ","를 없애야 한다!! 
        
        diff = tr.find_element(By.CSS_SELECTOR,".rank_wrap").get_attribute("title")
        diff_text = ""
        if "순위동일" in diff:
            diff_text = "-"
        elif "상승" in diff:
            diff_text = "+" + tr.find_element(By.CSS_SELECTOR,".rank_wrap .up").text
        elif "하락" in diff:
            diff_text = "-" + tr.find_element(By.CSS_SELECTOR,".rank_wrap .down").text
        elif "진입" in diff:
            diff_text = "new"
        rank = rank.text.replace("'", "")
        title = title.text.replace("'", "")
        singer = singer.text.replace("'", "")
        album = album.text.replace("'", "")
        like = like.text.replace(",","")
        
        print(rank)
        print(title)
        print(singer)
        print(album)
        print(like)
        print(diff_text)
        
        sql = f""" 
        INSERT INTO melon 
        VALUES
            (NULL, {rank}, '{title}', '{singer}', '{album}', {like}, NULL);

        """
        cursor.execute(sql)  

    db.commit()
    db.close()
    
except Exception as e:            # except는 필수! 오류났을 경우 error 메시지 출력
    print(e)
finally:                          # finally는 선택사항 error가 날지도 모르는 사항
    time.sleep(2)
    d.close()
    d.quit()
             
```
