```python
# 
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from openpyxl import Workbook

wb = Workbook()  # workbook의 개념은 파일
ws = wb.active   # worksheet이 여러개 있다. active는 생성했을 때 그 sheet를 선택한 것.

service = Service()
d = webdriver.Chrome(service=service)
try:
    d.get("https://cafe.naver.com/joonggonara.cafe?iframe_url=/ArticleList.nhn%3Fsearch.clubid=10050146%26search.boardtype=L%26viewType=pc")
    
    elem = d.find_element(By.CSS_SELECTOR, "#topLayerQueryInput")
    elem.send_keys("자전거")
    elem.send_keys(Keys.RETURN)
    
    # 대기 시간을 주어 페이지가 로드되도록 합니다.
    time.sleep(2)
    
    iframe = d.find_element(By.CSS_SELECTOR, "#cafe_main")
    d.switch_to.frame(iframe)
    
    # 원본 웹으로 돌아오는 함수
    # d.switch_to.default_content()
    
    for i in range (2,7): 
        
#     area = d.find_elements(By.CSS_SELECTOR, ".article-board")[1] -> article-board가 한개가 아니다! 따라서 elements
#     trs = area.find_elements(By.CSS_SELECTOR, "table>tbody>tr")
#     위처럼 하는 것보다 아래처럼 하는게 더 편하다.

        trs = d.find_elements(By.CSS_SELECTOR, ".article-board > table > tbody > tr")
    
        for tr in trs:
            article = tr.find_element(By.CSS_SELECTOR, ".article")
            print(article.text)
            name = tr.find_element(By.CSS_SELECTOR, ".p-nick > a")
            print(name.text)
            date = tr.find_element(By.CSS_SELECTOR, ".td_date")
            print(date.text)
            ws.append([article.text, name.text, date.text])
  
        page_area = d.find_element(By.CSS_SELECTOR, ".prev-next")
        page_link = page_area.find_element(By.LINK_TEXT,str(i))
        page_link.click()
        
    
                  
except Exception as e:
    print(e)
finally:
    time.sleep(2)
    d.close()
    d.quit()
    
wb.save("Joongo.xlsx")
```
