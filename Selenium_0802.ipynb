```python
# selenium은 자동화
# chrome은 최적화를 많이 해준다. 각 기기마다 그래픽카드, 블루투스 등등을 다 확인하고 설치해준다. ->굉장히 부드럽다.
# chromedriver를 사용한다면 그런것 없다.
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service(ChromeDriverManager().install())
d = webdriver.Chrome(service=service)

try:
    d.get("https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105")
    #elem = d.find_element(By.CSS_SELECTOR, "#_rankingList0")
    #titles = elem.find_elements(By.CSS_SELECTOR, ".list_tit")
    #print(elem.text)
    #for title in titles:
    #    print(title.text)
    
    time.sleep(0.5)
    
    section = d.find_element(By.CSS_SELECTOR, ".section_body")
    lis = section.find_elements(By.CSS_SELECTOR, "ul>li")
    for li in lis:
        print(li.text)
        
    page_area = d.find_element(By.CSS_SELECTOR, "#paging")
    page_2 = page_area.find_element(By.LINK_TEXT,"2")
    page_2.click()
    
    time.sleep(2)
    
    section = d.find_element(By.CSS_SELECTOR, ".section_body")
    lis = section.find_elements(By.CSS_SELECTOR, "ul>li")
    for li in lis:
        print(li.text)
        
except Exception as e:            # except는 필수! 오류났을 경우 error 메시지 출력
    print(e)
finally:                          # finally는 선택사항 error가 날지도 모르는 사항
    time.sleep(2)
    d.close()
    d.quit()

```

```python
# 반복문 이용해서 1-5페이지까지 출력하기
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service()
d = webdriver.Chrome(service=service)

try:
    d.get("https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105")
    #elem = d.find_element(By.CSS_SELECTOR, "#_rankingList0")
    #titles = elem.find_elements(By.CSS_SELECTOR, ".list_tit")
    #print(elem.text)
    #for title in titles:
    #    print(title.text)     
    
    time.sleep(0.5)
    
    for i in range (2,6):  #question!: (1,6) 아닌 이유 -> 애초에 get할때 1페이지 상태이다.
        
        section = d.find_element(By.CSS_SELECTOR, ".section_body")
        lis = section.find_elements(By.CSS_SELECTOR, "ul>li")
        for li in lis:
            print(li.text)

        page_area = d.find_element(By.CSS_SELECTOR, "#paging")
        page_link = page_area.find_element(By.LINK_TEXT,str(i))
        page_link.click()

        time.sleep(2)
        print("-"*30, i, "-"*30)
        
except Exception as e:            # except는 필수! 오류났을 경우 error 메시지 출력
    print(e)
finally:                          # finally는 선택사항ㅉ error가 날지도 모르는 사항
    time.sleep(2)
    d.close()
    d.quit()

```
