# 계단 오르내리기 예제

```python
import random
import matplotlib.pyplot as plt

position = 0
walk=[position]
steps=1000

for i in range(steps):
    step=1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

```python
import numpy as np

nsteps =1000
rng = np.random.default_rng(seed=12345)
draws = rng.integers(0,2,size=nsteps)

steps=np.where(draws == 0,1,-1)

walk = steps.cumsum()

walk
```

```python
walk.max()
```

```python
(np.abs(walk) >=10).argmax()
# 걷기가 10에 도달하거나 초과한 위치를 나타내는 부울 배열 제공
# 부울 배열에서 최댓값의 첫 번째 인덱스를 반환하는 룰 사용하여 계산
```

```python
nwalks = 5000
nsteps = 1000
draws = rng.integers(0,2,size=(nwalks,nsteps))
steps = np.where(draws >0,1,-1)
walks = steps.cumsum(axis =1)  # axis=1은 행 방향으로 누적 합계를 계산하라는 의미
walks
```

```python
walks.max()
```

```python
hits30 = (np.abs(walks) >= 30).any(axis = 1)
hits30
hits30.sum()
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis = 1)
crossing_times
```

```python
crossing_times.mean()
```
